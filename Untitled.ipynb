```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Create a graph object
G = nx.Graph()

# Add nodes with random heuristic values
for node in range(15):
    heuristic_value = random.randint(1, 15)
    G.add_node(node, heuristic=heuristic_value)

# Create edges to ensure the graph is connected
edges = [(0, 1, 3), (0, 2, 2), (1, 3, 4), (1, 4, 5), (2, 5, 6), (2, 6, 7), (3, 7, 8), (4, 8, 9), (5, 9, 10), (6, 9, 11)]
G.add_weighted_edges_from(edges)

# Add more random edges to increase connectivity
for _ in range(15):
    node1 = random.randint(0, 14)
    node2 = random.randint(0, 14)
    while node1 == node2 or G.has_edge(node1, node2):
        node1 = random.randint(0, 14)
        node2 = random.randint(0, 14)
    weight = random.randint(1, 10)
    G.add_edge(node1, node2, weight=weight)

# Get positions for nodes (you can customize this)
pos = nx.spring_layout(G)

# Draw nodes with labels and heuristic values
nx.draw(G, pos, with_labels=True, labels={node: f"H: {G.nodes[node]['heuristic']}" for node in G.nodes}, node_size=1000, node_color='lightblue')

# Draw edges with weights
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
nx.draw_networkx_edges(G, pos)

# Display the graph
plt.show()
```
